```python
import os
import mne
import numpy as np
import functions as f
```

```python
#for 16 channels! 
def transform_data_16(dataDir, saveDir, sliding_window_size, sliding_window_overlap, z_normalize=True):
	# iterate over files in
	# that directory
	os.makedirs(saveDir, exist_ok=True)
	for i,filename in enumerate(os.scandir(dataDir)):
		if filename.is_file():
			raw_h = mne.io.read_raw_edf(filename.path, preload=True)
			raw_h.drop_channels(
                    [
                        "EEG A2-A1",
                        "EEG 23A-23R",
                        "EEG 24A-24R",
                        "EEG T6-LE",
                        "EEG Cz-LE",
                        "EEG Pz-LE",
                    ],
                    on_missing="ignore",
                )
			raw_h = raw_h.rename_channels(
                    {
                        "EEG Fp1-LE": "Fp1",
                        "EEG F3-LE": "F3",
                        "EEG C3-LE": "C3",
                        "EEG P3-LE": "P3",
                        "EEG O1-LE": "O1",
                        "EEG F7-LE": "F7",
                        "EEG Fz-LE": "Fz",
                        "EEG Fp2-LE": "Fp2",
                        "EEG F4-LE": "F4",
                        "EEG C4-LE": "C4",
                        "EEG P4-LE": "P4",
                        "EEG O2-LE": "O2",
                        "EEG F8-LE": "F8",
                        "EEG T3-LE" : "T3",
                        "EEG T5-LE" : "T5",
                        "EEG T4-LE" : "T4",
                    }
                )
			raw_filter = f.filter_raw(raw_h, reference='average')
			raw_montage = f.make_montage(raw_filter)
			ica_filter = f.ica_filter(raw_montage,n_components=0.99)
			data_array = f.sliding_window_eeg_manually(ica_filter,sliding_window_size, sliding_window_overlap, z_normalize=z_normalize, normalize=False)
			#save the data_array
			np.save(saveDir + '/' + filename.name[:-4], data_array)
```

```python
#for 16 channels! 
def transform_data_13(dataDir, saveDir, sliding_window_size, sliding_window_overlap, z_normalize=True):
	# iterate over files in
	# that directory
	os.makedirs(saveDir, exist_ok=True)
	for i,filename in enumerate(os.scandir(dataDir)):
		if filename.is_file():
			raw_h = mne.io.read_raw_edf(filename.path, preload=True)
			raw_h.drop_channels(['EEG A2-A1', 'EEG 23A-23R', 'EEG 24A-24R', 'EEG T3-LE', 'EEG T5-LE', 'EEG T4-LE', 'EEG T6-LE', 'EEG Cz-LE', 'EEG Pz-LE'], on_missing='ignore')
			raw_h = raw_h.rename_channels({'EEG Fp1-LE': 'Fp1', 'EEG F3-LE': 'F3', 'EEG C3-LE': 'C3', 'EEG P3-LE': 'P3', 'EEG O1-LE': 'O1', 'EEG F7-LE': 'F7', 'EEG Fz-LE': 'Fz', 'EEG Fp2-LE': 'Fp2', 'EEG F4-LE': 'F4', 'EEG C4-LE': 'C4', 'EEG P4-LE': 'P4', 'EEG O2-LE': 'O2', 'EEG F8-LE': 'F8'})
			raw_filter = f.filter_raw(raw_h, reference='average')
			raw_montage = f.make_montage(raw_filter)
			ica_filter = f.ica_filter(raw_montage,n_components=0.99)
			data_array = f.sliding_window_eeg_manually(ica_filter,sliding_window_size, sliding_window_overlap, z_normalize=z_normalize, normalize=False)
			#save the data_array
			np.save(saveDir + '/' + filename.name[:-4], data_array)
```

```python
def edf_to_array(data_dir, save_dir, sliding_window_size=4, sliding_window_overlap=0.5, z_normalize=True):
	#load and preprocess the data for Healthy
	dataDir_H = os.path.join(data_dir,"H_EC")
	saveDir_H = os.path.join(save_dir,"H_EC")
	os.makedirs(saveDir_H, exist_ok=True)
	f.remove_files_in_directory(saveDir_H)
	transform_data_13(dataDir=dataDir_H, saveDir=saveDir_H, sliding_window_size=sliding_window_size, sliding_window_overlap=sliding_window_overlap, z_normalize=z_normalize)

	#load and preprocess the data for MDD
	dataDir_MDD = os.path.join(data_dir,"MDD_EC")
	saveDir_MDD = os.path.join(save_dir,"MDD_EC")
	os.makedirs(saveDir_MDD, exist_ok=True)
	f.remove_files_in_directory(saveDir_MDD)
	transform_data_13(dataDir=dataDir_MDD, saveDir=saveDir_MDD, sliding_window_size=sliding_window_size, sliding_window_overlap=sliding_window_overlap, z_normalize=z_normalize)
```
